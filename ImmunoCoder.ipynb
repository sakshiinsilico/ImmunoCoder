### PCR Simulation
This simulates exponential DNA amplification during PCR cycles.

```python
import numpy as np
import matplotlib.pyplot as plt

cycles = np.arange(1, 31)
dna = 2 ** cycles

plt.plot(cycles, dna)
plt.xlabel("PCR Cycles")
plt.ylabel("DNA Quantity")
plt.title("PCR Amplification Simulation")
plt.show()
```

### PCR Simulation
This simulates exponential DNA amplification during PCR cycles.

```python
import numpy as np
import matplotlib.pyplot as plt

time = np.arange(0, 20, 1)
response = time * np.exp(-0.3*time)

plt.plot(time, response)
plt.xlabel("Time")
plt.ylabel("Immune Response")
plt.title("Immune Response Simulation")
plt.show()
```

### FACS Simulation
This simulates fluorescence intensity distribution of cell populations measured using flow cytometry (FACS).

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

negative_cells = np.random.normal(loc=200, scale=50, size=1000)
positive_cells = np.random.normal(loc=800, scale=100, size=1000)

plt.axvline(500, color='red', linestyle='--', label="Gating Threshold")
plt.hist(negative_cells, bins=50, alpha=0.6, label="Negative Cells")
plt.hist(positive_cells, bins=50, alpha=0.6, label="Positive Cells")

plt.xlabel("Fluorescence Intensity")
plt.ylabel("Cell Count")
plt.title("FACS Simulation: Cell Population Distribution")
plt.legend()
plt.show()
```
